```python
plt.figure(figsize=(10,10))

plt.plot(controlT,std['CN'])
```

```python
ax1 = plt.subplot(1,3,1)
plt.plot(guiltT,guilt['CG'], ax = ax1)

ax2 = plt.subplot(1,3,2)
plt.plot(controlT,control['CN'] , ax = ax2)

ax3 = plt.subplot(1,3,3)
plt.plot(controlT,std['CN'] , ax = ax3)
```
